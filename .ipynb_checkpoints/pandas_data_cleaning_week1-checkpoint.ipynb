```python
import pandas as pd

# Step 1: Load dataset
file_path = "jeans.csv"  

df = pd.read_csv(file_path)
print("Dataset loaded successfully")

# Step 2: Explore dataset
print("\nFirst 5 rows")
display(df.head())
print("\nLast 5 rows")
display(df.tail())

print("\nShape:", df.shape)

print("\nColumns:", df.columns)

print("\nData types:")
print(df.dtypes)

# Step 3: Clean final_price properly
df['final_price'] = (
    df['final_price']
    .astype(str)
    .str.replace(r'[‚āĻ,"]', '', regex=True)
    .str.strip()
)

df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')

# Step 4: Handle missing values
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating'] = df['rating'].fillna(df['rating'].median())

df['discount'] = pd.to_numeric(df['discount'], errors='coerce')
df['discount'] = df['discount'].fillna(0)

print("\nMissing values handled")

# Step 5: Basic analysis
print("\nSelected columns")
display(df[['final_price', 'rating']].head())

print("\nProducts with rating > 4")
display(df[df['rating'] > 4].head())

# Step 6: Remove duplicates
before = df.shape
df = df.drop_duplicates()
after = df.shape

print("\nBefore duplicates:", before)
print("After duplicates:", after)

# STEP 7: DERIVED COLUMN + INSIGHTS + SAVE
df['quantity'] = 2

df['final_price'] = df['final_price'].fillna(0)
df['total_amount'] = df['final_price'] * df['quantity']

print("\nTotal amount sample")
display(df[['final_price', 'quantity', 'total_amount']].head())

print("\nBasic insights")
print("Average rating:", df['rating'].mean())
print("Max price:", df['final_price'].max())
print("Min price:", df['final_price'].min())

df.to_csv("cleaned_shopping_data.csv", index=False)
print("\nCleaned dataset saved successfully")


# SUMMARY

print("""
FINAL SUMMARY:

- Worked on a shopping dataset (jeans.csv) using Pandas for data cleaning and analysis.

- Loaded the dataset and explored its structure using head(), tail(), shape, columns, and data types.

- Cleaned the final_price column by removing currency symbols and converting it into numeric format.

- Handled missing values in rating and discount columns by filling with median and 0 respectively.

- Removed duplicate records to ensure data accuracy.

- Created a new feature column called total_amount (final_price √ó quantity).

- Performed filtering to display products with rating greater than 4.

- Generated basic insights such as average rating, maximum price, and minimum price.

- Saved the cleaned dataset as 'cleaned_shopping_data.csv' for further use.
""")
```
